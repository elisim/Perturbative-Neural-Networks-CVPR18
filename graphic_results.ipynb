```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
 
objects = ('MNIST', 'EMNIST', 'CIFAR-10', 'CIFAR-100')
y_pos = np.arange(len(objects))
accuracy = [98.22,85.39,86.40,55.00]

plt.figure(figsize=(7, 5))  # width:20, height:3 
plt.bar(range(len(accuracy)), accuracy, align='edge', width=0.1,color='b')
plt.xticks(y_pos, objects)
plt.ylabel('Accuracy')
plt.title('Accuracy per Dataset')
 
plt.show()
```
